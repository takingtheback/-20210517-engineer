<데이터 전처리>

```python
# null 데이터 확인
df.isnull().sum()
```

```python
x_data = df[['cap-shape', 'cap-color', 'gill-size', 'gill-color', 'stalk-shape']]
x_data 
```

```python
# y_data 
y_data = df['class']
y_data
```

```python
# 문자 데이터 원-핫 인코딩
y_data = pd.get_dummies(y_data, columns=['class'])
y = y_data['p']
y
```

```python
x = pd.get_dummies(x_data)
x
```

```python
# gill-size 하나만 남기기
x = x.drop(['gill-size_n' ], axis=1)
x
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

x = torch.Tensor(x.values)
x.shape
```

```python
y = torch.Tensor(y)
y = y.unsqueeze(1)
y.shape
```

```python
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

X_train = torch.Tensor(X_train)
X_test = torch.Tensor(X_test)
y_train = torch.Tensor(y_train)
y_test = torch.Tensor(y_test)

ds_train = TensorDataset(X_train, y_train)
ds_test = TensorDataset(X_test, y_test)

loader_train = DataLoader(ds_train, batch_size=60, shuffle=True)
loader_test = DataLoader(ds_test, batch_size=60, shuffle=False)
```

<모델 생성>

```python
model = nn.Sequential()

model.add_module('fc1', nn.Linear(31, 200))
model.add_module('relu1', nn.ReLU()) 

model.add_module('fc2', nn.Linear(200, 200))
model.add_module('relu2', nn.ReLU())

model.add_module('fc3', nn.Linear(200, 200))
model.add_module('relu3', nn.ReLU())

model.add_module('fc4', nn.Linear(200, 1))
model.add_module('sig', nn.Sigmoid())

print(model)
```

<비용함수, 옵티마이저 생성>

```python
loss_fn = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

<학습>

```python
model.train()

for e in range(31):
    for data, targets in loader_train:

        optimizer.zero_grad()  
        outputs = model(data)  
        loss = loss_fn(outputs, targets)  
        loss.backward()  
        optimizer.step()
    if e%10==0:
        print(e, '/30', ',', 'cost:', loss.item())
```

<테스트>

```python
model.eval()  
correct = 0

with torch.no_grad():  
    for data, targets in loader_test:

        outputs = model(data)  
        
        predicted = (outputs.data > 0.5).float()
        correct += predicted.eq(targets.data.view_as(predicted)).sum()  

data_num = len(loader_test.dataset)  
print('\n테스트 데이터에서 예측 정확도: {}/{} ({:.0f}%)\n'.format(correct, data_num, 100. * correct / data_num))
```
